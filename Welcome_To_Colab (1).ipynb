 2. Explain the difference between mutable and immutable data types with examples.
  - Mutable data types are those data types whose values can be changed after they are created.

Immutable data types are those data types whose values cannot be changed once they are created.

Mutable Data Types

Data can be modified

Changes affect the same object

Examples:

List
Example:
marks = [80, 85, 90]
If we change one value:
marks[0] = 95
New list → [95, 85, 90]

Dictionary
Example:
student = {"name": "Aman", "class": 9}
We can change the class to 10.

Immutable Data Types

Data cannot be modified

A new object is created if we try to change it

Examples:

Integer
Example:
x = 10
x = 20 (old value 10 is not changed, new value is assigned)

String
Example:
name = "Ravi"
We cannot change "Ravi" to "Ravi Kumar" directly.

List

5. Why might you use a set instead of a list in Python.
 - I might use a set instead of a list in Python when I want to store unique values and don't care about the order.

For example, a set automatically removes duplicate elements, while a list allows duplicates. Also, checking whether an element exists in a set is faster than in a list.
So, sets are useful when I need:
No repeated items
Faster searching
Just a collection of values, not in a specific order
That's why a set can be better than a list in some situations.

7. How do tuples ensure data integrity in Python?
 - Tuples help ensure data integrity in Python because they are immutable, which means their values cannot be changed after they are created.
Once a tuple is made, no items can be added, removed, or modified. This prevents accidental changes to important data. Because of this, tuples are often used when data should remain fixed and safe throughout the program.


8. What is a hash table, and how does it relate to dictionaries in Python ?
 - A hash table is a data structure that stores data in key-value pairs and allows fast searching, inserting, and deleting.
In Python, a dictionary uses a hash table internally. Each key is passed through a hash function, which decides where the key-value pair is stored in memory. Because of this, dictionaries can quickly find values using their keys.
So, dictionaries are based on hash tables, which is why accessing data in a dictionary is very fast.

9.  Can lists contain different data types in Python?
 - Yes, lists can contain different data types in Python.
A single list can store integers, strings, floats, and even other lists together. For example, a list can have a number, a word, and another list inside it.
This makes lists very flexible and useful for storing mixed types of data in one place.

10.  Explain why strings are immutable in Python.
 - Strings are immutable in Python to keep them safe and efficient to use.
Once a string is created, it cannot be changed. If we try to modify it, Python actually creates a new string instead of changing the old one. This helps prevent accidental changes and makes strings faster to handle in memory, especially when the same string is used in many places.

So, dictionaries are better than lists when we need quick lookup and clear relationships between data.

12. Describe a scenario where using a tuple would be preferable over a list.
 - Using a tuple is preferable over a list when the data should not be changed.

13. How do sets handle duplicate values in Python?
 - In Python, sets do not allow duplicate values.
If we try to add the same value more than once to a set, Python automatically keeps only one copy of that value and removes the duplicates. This happens because sets are designed to store unique elements only.


14.  How does the “in” keyword work differently for lists and dictionaries ?
 - The in keyword is used to check whether something exists in a collection, but it works a little differently for lists and dictionaries.
For a list, in checks whether a value is present anywhere in the list.
For a dictionary, in checks only the keys, not the values.

17. Describe the time complexity of accessing elements in a dictionary.
 - Accessing elements in a dictionary is very fast because dictionaries use a hash table internally.
The time complexity for looking up a value using its key is usually O(1), which means it takes constant time no matter how big the dictionary is.
This is much faster than searching in a list, which can take O(n) time in the worst case.

18. In what situations are lists preferred over dictionaries?
 - Lists are preferred over dictionaries when we need:
Ordered data - Lists keep items in a specific order, while dictionaries don't guarantee order (though Python 3.7+ preserves insertion order, lists are still simpler for sequences).
Duplicate values - Lists can store repeated items, but dictionaries require unique keys.
Simple collections - If we just want a collection of items without needing key-value pairs, a list is easier to use.
Index-based access - Lists let us access elements by their position using indices, which dictionaries don't support.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
 - Dictionaries are considered unordered because, in general, the items are not stored in a specific sequence—their position doesn’t depend on the order in which we add them.
This means we cannot access items by their position or index like we do in a list. Instead, we must use the key to retrieve a value.

20.  Explain the difference between a list and a dictionary in terms of data retrieval.
 - The main difference between a list and a dictionary in terms of data retrieval is how we access the data:
In a list, we access items by their index (position), like my_list[2]. Searching for a value can take longer if the list is big because Python may need to check many items.
In a dictionary, we access items by their key, like my_dict["name"]. This is usually much faster because dictionaries use a hash table, so they can find the value almost instantly.
