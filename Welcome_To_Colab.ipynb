**Theory Question **

1. What is Python, and why is it popular?
 - Python is a high-level programming language that is easy to understand and use.
It is popular because it has simple syntax, is beginner-friendly, and is used in many areas like web development, data science, and AI.

2. What is an interpreter in Python?
- An interpreter in Python is a program that reads and runs Python code line by line.
It checks each line for errors and executes it immediately, which makes debugging easier.

3. ¬8 What are pre-defined keywords in Python ?
 - Pre-defined keywords in Python are special words that have a fixed meaning in the language.
They are used to write programs and cannot be used as variable names, such as if, for, and while.

4. Can keywords be used as variable names ?
 - No, keywords cannot be used as variable names.
This is because keywords have special meanings in Python, so using them as variable names would cause errors.

5. What is mutability in Python ?
 - Mutability in Python means whether we can change the value of an object after it is created.
If an object can be changed, it is called mutable (like lists), and if it cannot be changed, it is called immutable (like strings and tuples).

6. Why are lists mutable, but tuples are immutable ?
 - Lists are mutable because we can change, add, or remove their elements after creating them.
Tuples are immutable because once they are created, their values cannot be changed, which makes them more secure and faster to use.

7. What is the difference between “==” and “is” operators in Python?
 - The == operator is used to check if the values of two variables are the same.
The is operator is used to check if both variables refer to the same object in memory.

8. What are logical operators in Python ?
 - Logical operators in Python are used to combine or compare conditions.
The main logical operators are and, or, and not, and they return True or False based on the conditions.

9. What is type casting in Python ?
 - Type casting in Python means converting one data type into another data type.
For example, we can change an integer into a float or a string using functions like int(), float(), and str().

10. What is the difference between implicit and explicit type casting ?
Ans - Implicit type casting happens automatically by Python when there is no risk of data loss.
Explicit type casting is done by the programmer using functions like int(), float(), or str().

11. What is the purpose of conditional statements in Python ?
 - The purpose of conditional statements in Python is to make decisions in a program.
They allow the program to run different code based on conditions using statements like if, elif, and else.

12. How does the elif statement work ?
 - The elif statement is used to check another condition if the previous if condition is false.
It helps us test multiple conditions one by one in a program.
13. What is the difference between for and while loops ?
 - A for loop is used when we know how many times the loop should run.
A while loop is used when we do not know the number of times in advance and it runs until a condition becomes false.

14. Describe a scenario where a while loop is more suitable than a for loop.
 - A while loop is more suitable when we do not know in advance how many times the loop will run.
For example, when taking input from a user until they enter the correct password, a while loop is better to use.

```
# 6. Write a function to demonstrate the behavior of mutable and immutable arguments.
# Function with immutable argument
def change_number(num):
    num = num + 10
    print("Inside function (immutable):", num)

# Function with mutable argument
def change_list(lst):
    lst.append(10)
    print("Inside function (mutable):", lst)

# Main program
number = 5
numbers_list = [1, 2, 3]

change_number(number)
print("Outside function (immutable):", number)

change_list(numbers_list)
print("Outside function (mutable):", numbers_list)
```

```
# 9. Write a Python program to convert user input from string to integer, float, and boolean types.
# Take input from user (input is always a string)
user_input = input("Enter something: ")

# Convert to integer
try:
    int_value = int(user_input)
    print("Integer:", int_value)
except ValueError:
    print("Cannot convert to integer.")

# Convert to float
try:
    float_value = float(user_input)
    print("Float:", float_value)
except ValueError:
    print("Cannot convert to float.")

# Convert to boolean
# In Python, empty string '' is False, anything else is True
bool_value = bool(user_input)
print("Boolean:", bool_value)
```

```
# 10. Write code to demonstrate type casting with list elements.
# Original list with mixed string numbers
original_list = ["10", "20", "30", "40"]
print("Original list (strings):", original_list)

# Convert each element from string to integer
int_list = [int(x) for x in original_list]
print("List after converting to integers:", int_list)

# Convert each element from integer to float
float_list = [float(x) for x in int_list]
print("List after converting to floats:", float_list)

# Convert each element back to string
str_list = [str(x) for x in float_list]
print("List after converting back to strings:", str_list)
```

```
# 13. Write a Python program to find the sum of all even numbers between 1 and 50.
# Initialize sum
even_sum = 0

# Loop through numbers from 1 to 50
for i in range(1, 51):
    if i % 2 == 0:  # Check if the number is even
        even_sum += i

# Display the result
print("Sum of all even numbers between 1 and 50 is:", even_sum)
```

```
# 14. Write a program to reverse a string using a while loop.
# Input from user
string = input("Enter a string: ")

# Initialize variables
reversed_string = ""
index = len(string) - 1  # Start from the last character

# Loop to reverse the string
while index >= 0:
    reversed_string += string[index]
    index -= 1  # Move to the previous character

# Display the reversed string
print("Reversed string:", reversed_string)
```

```
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
# Input from user
num = int(input("Enter a number: "))

# Initialize factorial and counter
factorial = 1
counter = 1

# Calculate factorial using while loop
while counter <= num:
    factorial *= counter
    counter += 1

# Display the result
print(f"Factorial of {num} is: {factorial}")
```
